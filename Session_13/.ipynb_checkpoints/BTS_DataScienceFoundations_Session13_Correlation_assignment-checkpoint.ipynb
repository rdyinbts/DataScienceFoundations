```python
df.drop(['PassengerId','Name','Ticket','Cabin','Embarked'], axis=1).corr(method='spearman')
```

```python
df.drop(['PassengerId','Name','Ticket','Cabin','Embarked'], axis=1).corr(method='spearman').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'),axis=1)
```

```python
corr = df.drop(['PassengerId','Name','Ticket','Cabin','Embarked'], axis=1).corr(method='spearman')

mask = np.zeros_like(corr,dtype=np.bool)
mask[np.triu_indices_from(mask)]=True

f,ax = plt.subplots(figsize=(11,9))
cmap= sns.diverging_palette(220,10, as_cmap=True)

sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,square=True, linewidths=.4, cbar_kws={"shrink":.5})
```
