```python
import sqlite3
from sqlite3 import Error
import pandas as pd
```

```python
def create_connection(db_file):
    """ 
    Create a database connection to the SQLite database specified by the db_file
    
    :param db_file: database file
    :return: the Connection object or None
    """
    try:
        conn = sqlite3.connect(db_file)
        return conn
    except Error as e:
        print(e)
 
    return None
```

```python
# SUBSTITUTE THIS PATH BY THE CORRECT PATH ON YOUR LOCAL FILE SYSTEM
database = "/Users/racheldyap/Desktop/DataScienceFoundations/Session_3/Titanic.db"

# Create a database connection
conn = create_connection(database)

# explicitly state number of rows to display at N=10
N = 10

#you query using SQL 'codes'
# here you select * \all data\ from your SQL table passengers
# and put code to indicate # of rows you want it to display
# using 'LIMIT {}.format(N)' where N is equal to 10
query = "SELECT * FROM passengers LIMIT {}".format(N)

#assign to a Pandas dataframe to read sql
# give parameters of 1 QUERY and 2 Connector
df = pd.read_sql_query(query, conn)

#close the connector after each query
conn.close()

df.head()
```
