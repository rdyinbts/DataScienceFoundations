## Finding outliers

```python
titanic['age_std'] = ((titanic['Age'] - titanic['Age'].mean()).apply(abs)/titanic.Age.std())
```

```python
titanic.sort_values(by='age_std', ascending=True)
```

```python
## there are a lot of people whose ages are outside the 1.5 acceptable standard deviation from the mean 
```

## Skewness and Kurtosis

```python
titanic.Age.plot(kind='box', whis='range')
```

```python
# take out outliers for age (those with std less than 1.5)
```

## ? how can i take out std > and < 1.5 in one command line

```python
age_cleaned_titanic = titanic[titanic['age_std']< 1.5]
```

```python
age_cleaned_titanic.sort_values(by='age_std', ascending=False)
```

```python
age_cleaned_titanic.shape
```
